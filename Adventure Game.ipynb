```python
"""
Created on Thu January 31 17:00 2023
@author: danielevenschor
"""
"""
    DocString:
    
    A) Introduction:
    This game is based on the film series Lord of the Rings / Hobbit.
    It consists of three main stages.
    Having watched the films will help in some cases.
    
    Stage 1: The Shire
    Stage 2: Mines of Moria
    Stage 3: Mordor / Mount Doom
    
    B) Defined Functions:
    FAIL() 
    AGAIN()
    Stage0()
    Stage1()
    Stage2()
    TIMER()
    ATTACKSPOT()
    WATCHER()
    BALROG()
    STAGE3()
    FINAL()

    C) Known "cheats":
    You can skim through the choices in BALROG with invalid inputs to re-roll what he does. 
    This makes it very easy to pass. 
    
    
"""

import random # random will be used to make the outcomes change for every playthrough
from sys import exit # will be used to fail the game
import time # will be used for the timer function
power = 1 # set powers to 1 (this is for the ring)

def FAIL(stage):
    
    """
    DOCSTRING:
    ------------------------------------------------------------------------------------------------
    Purpose: Calling this function will trigger different prints depending on the stage you failed in. 
    ------------------------------------------------------------------------------------------------
    VARIABLES: 
    Stage: 
    0: 
    1:
    1_2:
    2:
    2_1:
    3:
    3_1:
    """

    if stage == 0:
        print("""You decided not to take on the quest. 
Your journey has ended before it began.
You give back the ring to Bilbo and stay in your comfort zone.
-------------------------------------------------------------
Probably you would have died anyways, coward!
Maybe one day someone will take on the challenge.
... """)
        AGAIN() 
    elif stage == 1:
        print("""
You are just a tiny little hobbit, not a warrior. 
The orcs capture you and find the ring.
Not only are you going to die, you also gave the orcs what they have been looking for.
You are a failure!

GAME OVER!
""")
        AGAIN()
    elif stage == 1_2:
        print("""
You tried to run, but tripped over a root in the forest and broke your leg.
Unfortunately the orcs catch up to you and kill you.
Today was not your lucky day...

GAME OVER!
""")
        AGAIN()
    elif stage == 2:
        print("""You died to the Watcher of the Water. What a joke!
You would not have made it far anyway with such fighting skills.

GAME OVER!
""")
        AGAIN()

    elif stage == 2_1:
        print("""You died to the Balrog. You were not careful enough im afraid!
No chance to fight, he knocks you off the bridge with his fire whip and your journey has ended.

R.I.P
""")
        AGAIN()
    elif stage == 3:
        print("""You have attempted to sneak past the gate.
Unfortunately, you were spotted by an enemy and captured!

GAME OVER!
""")

        AGAIN()
    elif stage == 3_1:
        print("""You have kept and used the ring for yourself.
Unfortunately, you were quickly overcome by its power and became a servant of Sauron!

GAME OVER!
""")
        AGAIN()

def AGAIN():

    """
    DOCSTRING:
    ----------------------------------------------------------------------------------
    Purpose: Calling this function triggers the play again option, restarting the game
    Arguments: None needed
    ----------------------------------------------------------------------------------
    """

    play = input("\n\nDo you want to play again? [yes/no] ").lower()
    while play not in ["yes", "no"]: # while loop to control user input
        play = input("""\nInvalid input! 
Do you want to play again? Make sure to answer "Yes" or "No": """).lower()
    if play == "yes":
        global power
        power=1 # reset the power to 1 in case it was used in the first playthrough
        STAGE1() # Start the first stage
    else:
        exit(0) # exit the game

def STAGE0(): # Stage 0: Introduction
    # documenting the function
    """
    DOCSTRING:
    ------------------------------------------
    Calling this function will begin the game.
    No arguments needed. 
    ------------------------------------------
    """
    # Print the Introduction 
    print(f"""
Welcome to the Lord of the Rings Text Adventure Game! 

You will be playing Frodo, a young hobbit who has received a precious ring from his uncle Bilbo.
This ring allows you to become invisible, but it can only be used once in the game - so use it wisely!
You must take the ring to Mordor, where it has to be destroyed in Mount Doom.
On the way to Mount Doom you will face many difficult decisions that could end your journey. 
""")

# Stage 0: Make the decision to play or not
    d1 = input("""Do you accept the quest? ["Yes" or "No"]: """).lower()
    
# Controlling the user input for decision 1
    while d1 not in ["yes", "no"]:
        d1 = input("""Invalid input! 
Do you accept the quest? Make sure to answer "Yes" or "No": """).lower()

# Player decided to play the game:
    if d1 == "yes":
        print("""
You decided to take on the adventure, congratulations, you are not a coward!
Let's begin your journey, the first step is leaving your homeland.""")
        STAGE1() # start stage 1
    # The player decided not to play the game
    elif d1 == "no":
        FAIL(0) # call fail function message for this stage

def STAGE1(): # Stage 1: Start of the journey: The Shire
    global power # calling the global power variable, so it can be used in the function
    """ 
    DOCSTRING: 
    -------------------------------------------------------------
    Calling this function will start the first stage: "The Shire"
    No variables are required
    -------------------------------------------------------------
    """

    # The player faces decision number 2
    print("""Welcome to the Shire, the "peaceful" land of the Hobbits.
As you're travelling through the Shire, you come across a fork in the road that you have never seen before.
You have heard rumors of orcs invading the surrounding regions, you have to make a choice.""")
    d2 = input("""Do you go ["Left" or "Right]? """).lower()
    
    # Controlling user input for decision 2
    while d2 not in ["left", "right"]:
        d2 = input("""Invalid Choice! Please make sure to answer ["Left" or "Right"]. """).lower()
        
    # The player chose to go left
    if d2 == "left":
        print("""
You ran into a trap!
A group of orcs that set up a trap in a narrow passage want to skin and eat you alive.""")
        
        # The user faces their next decision, 2_1
        d2_1 = input("""What will you do? "fight", "run" or use your "ring" to become invisible to stay on track? """).lower() 
        # Controlling user output for decision 2_1 
        while d2_1 not in ["fight", "run", "ring"]: 
            d2_1 = input("""Something went wrong! Make sure to answer correctly! ["Fight", "Run" or "Ring"]
What will you do? "fight", "run" or use your "ring" to become invisible to stay on track? """).lower()
            
        # Player chose to Fight
        if d2_1 == "fight":
            FAIL(1) # Player failed the game
            
        # Player chose to run
        elif d2_1 == "run":
            chance = random.randint(1, 10) # Generate a random number from 1-10
            # Chance for success is 90% and failure is 10 % 
            if chance > 9: # Player got unlucky
                FAIL(1_2) # Player lost the game 
                
            elif chance <= 9: # Player got "lucky" and continues his journey
                print("""
You managed run away and hide in a small nearby cave until the orcs gave up.
Sometimes it's an advantage to be a tiny hobbit, the orcs couldn't follow you into the cave because of their size.
After a cold night in the cave, you continue your journey, moving closer to Mordor.
""")    
                input("< Press Enter to continue to the next stage>\n") # press enter to confirm, gives the game "structure"
                STAGE2() # continue to stage 2
                
        # Player chose to use his ring power!
        elif d2_1 == "ring":
            power -= 1 # since this is the first opportunity to use it there is no check necessary. Reduce the power to 0
            print("""
You have used the power of your ring, no more invisibility but you escaped the orcs.
You continue your journey, moving closer to Mordor
""")
            input("< Press Enter to continue to the next stage>\n")
            STAGE2() # move on to stage 2
            
    # Player chose to go on the "Right" path
    elif d2 == "right":
        # A list of three riddles
        Riddles = [
        "This thing all things devours: birds, beasts, trees, flowers, gnaws iron and bites steel,\ngrinds hard stones to meal, it slays king, ruins town and beats high mountain down.", #(Answer: Time)
        "Alive without breath, as cold as death: Never thirsty, ever drinking, all in mail never clinking.", #(Answer: Fish)
        "A box without hinges, key, or lid, yet golden treasure inside is hid." #(Answer: Egg)
        ]
        # Randomly select one of them to be asked 
        Riddle = random.choice(Riddles) 
        print(f"""
You come across a group of elves. They offer to help you on your journey if you can answer their riddle.
Their riddle is:"{Riddle}".\n[One word answer]""") 
        # The player now has to respond correctly, answers are commented in the list
        d2_1 = input("What is your answer? ").lower()
        if d2_1 == "time" and Riddle == Riddles[0]: # The answer is correct and player receives a hint from the elves
            print("""\nCorrect!
The elves are pleased with your intellect! 
Hearing that you are headed to Mordor, they tell you that the word "friend" opens a secret door into the Mines of Moria. 
They wish you the best of luck on your journey.
""") 
            input("< Press Enter to continue to the next stage>\n")
            STAGE2() #continue with stage 2
        elif d2_1 == "fish" and Riddle == Riddles[1]: # The answer is correct and player receives a hint
            print("""\nCorrect!
The elves are pleased with your intellect! 
Hearing that you are headed to Mordor, they tell you that the word "friend" opens a secret door into the Mines of Moria. 
They wish you the best of luck on your journey.
""") 
            input("< Press Enter to continue to the next stage>\n")
            STAGE2() 
        elif d2_1 == "egg" and Riddle == Riddles[2]: # The answer is correct and player receives a hint
            print("""\nCorrect!
The elves are pleased with your intellect! 
Hearing that you are headed to Mordor, they tell you that the word "friend" opens a secret door into the Mines of Moria. 
They wish you the best of luck on your journey.
""") 
            input("< Press Enter to continue to the next stage>\n")
            STAGE2() 
        else: # wrong answer but nothing happens the player doesnt receive a hint
            print("""\nIncorrect!
Don't worry, your journey continues. This was a hard riddle! 
You might have missed out on some valuable information from the elves but it't not a big deal.
The elves wish you safe travels and you continue on your journey. 
""") 
            input("< Press Enter to continue to the next stage>\n")
            STAGE2() # start stage 2

def STAGE2():
    """
    DOCSTRING:
    --------------------------------------------------------------
    Calling this function will begin Stage 2: "The Mines of Moria".
    No arguments needed. 
    --------------------------------------------------------------
    """

    global power # call the global "power" variable used for the ring
    print("""-----------------------------------------------------------------------
Welcome to the Mines of Moria.
You have successfully found the ancient door that leads into the Mines of Moria.

As you approach the door, you notice that it can only be opened with a riddle.

The riddle reads: "Speak, friend, and enter".\n""")
    attempts = 2 # set the ammounts of attempts, note that the first attempt is "free" so attempts=1 means you have 2 attempts 
    while attempts > 0:
        answer = input("What is the password to open the door? ").lower() # player has to solve the riddle -- ANSWER: "FRIEND"
        if answer == "friend": # correct answer
            attempts = -2 # set attempts to -2, this is used later as an argument
            print("""\n-------------------------------------------
WOW!! You solved the riddle! 
The door to the Mines of Moria opens with a loud creak.
""")       
            break # get out of the while loop 
        else: # wrong answer
            attempts -= 1
            print("Incorrect, try again!")
    if attempts == 0: # player failed 2 times
        print("""\n----------------------------------------------
You have guessed the wrong password two times... 
"Not that smart after all" ...
...
Suddenly, you hear water splashing in the lake right behind you.
A giant octopus-like creature stretches out of the water.
You must defeat the octopus to continue your journey. 
""") # introduce the upcoming fight
    if power>0 and attempts == 0: # possible option to use the ring and escape the situation
        answer = TIMER("""-------------------------------------------------------------------------
You still have not used your ring, this might be a good time to use it!
Type "ring" within 30 seconds to move on to the next stage:
Press Enter in order NOT to use the ring 
""", timeout=30) # timer has been used, the TIMER function is explained below!
        if answer == "ring": # player uses ring and moves on 
            power -= 1
            print("""You have made the decision to use your ring and dissappear, the octupus can't see you anymore.
You say "friend" and open the door to move on""")
            BALROG()
        else: # start the fight with the watcher in the water
            print("Time is up. Now you have to fight!")
            WATCHER()
    elif attempts == -2:
        BALROG() # start the game based on "attempts" declared earlier
    else: 
        WATCHER() 


def TIMER(prompt, timeout=15):
    """
    DOCSTRING:
    --------------------------------------------------------------------
    This function is used for timed inputs / prompts.
    --------------------------------------------------------------------
    Variables: 
    prompt: MANDATORY! Text string, explaining the desired user input
    timeout: OPTIONAL! Default: 15 seconds... 

    """
    start_time = time.time() # snapshot time at beginning
    input_string = input(prompt ).lower() # give the user time for his input
    end_time = time.time() # second snapshot of time
    if end_time - start_time > timeout: # take the difference and see if timer has been met or not
        return None
    else:
        return input_string
    
def ATTACKSPOT(): # This function is used in the WATCHER function to declare the point of attack
    """
    DOCSTRING:
    ----------------------------------------------------------------------------
    Calling this function generates a random choice of values from a dictionary
    No arguments needed. 
    ----------------------------------------------------------------------------
    """
    spot = {
        'q': 'left hand',
        'w': 'right hand',
        'e': 'left foot',
        'r': 'right foot'
    }

    return random.choice(list(spot.values())) # generate a random choice
    
def WATCHER():
    """
    DOCSTRING:
    -------------------------------------------------------------------------
    Calling this function will begin the battle with the Watcher in the water
    No arguments needed. 
    -------------------------------------------------------------------------
    """
    spot = { # need to declare the same dictionary as in "ATTACKSPOT" because we want to match the values with the keys
        'q': 'left hand',
        'w': 'right hand',
        'e': 'left foot',
        'r': 'right foot'
    }
    player_health = 10 # life points for this fight
    watcher_health = 100 

# Print the rules of the battle so the player understands what to do 
    print("""
The Watcher in the Water grabs your body parts with its tentacles, be ready to react quickly to cut of his tentacles.

You will have to cut off his tentacles a total of 4 times to win the fight!
If you mess up once, you die!
Here are the rules:

INPUT:
"Q" if he has your "LEFT HAND"
"W" if he has your "RIGHT HAND"
"E" if he has your "LEFT FOOT"
"R" if he has your "RIGHT FOOT

Note that you only have 10 seconds for each action, you might want to memorize the inputs!
GOOD LUCK!
----------------------------------------------------------------------------------------
""")
    input("< Press Enter when you are ready to start the fight >\n") # confirm with enter to start the fight
    while player_health > 0 and watcher_health > 0: # loop until either the player or the watcher is dead
        attack_spot = ATTACKSPOT()  # call ATTACKSPOT function, generating a random place of attack
        print(f"""\nThe Octopus grabs your {attack_spot.upper()}! You have 10 seconds to cut it off! 
""")  # print the place that the octopus grabs the player in UPPERCASE

        response = TIMER("\nQuick! Cut off the tentacle: ", timeout=10) # use of a TIMER, to give the input urgency

        if response == None:   # no response within 10 seconds 
            print("\nYou took too long and failed to cut off the tentacle")   
            player_health -= 10   # reduce player health by 10 -- You die!
            print("R.I.P you are dead")
            FAIL(2) # refers to the fail message of this stage

        elif response == list(spot.keys())[list(spot.values()).index(attack_spot)]:   # correct input according to dictionary keys 
            print("\nYou successfully cut off the tentacle")   
            watcher_health -= 25   # reduce watcher health by 25 each time, needs to happen 4 times to win the fight
            print(f""""\nThe Watcher's health is now {watcher_health}.\n """)

        else:   # incorrect input according to dictionary keys, decision not to control failed inputs because it has been explained in detail
            print("You failed to cut off the tentacle")  
            player_health -= 10   # player dies!
            print("R.I.P you are dead")
            FAIL(2)
    print("""\nYou defeated the Watcher of the Water! Close... You fought like a real hero!
Bilbo would be proud of you!
-------------------------------------------------------------------------------------------
""")
    input("< Press Enter to go through the now opened door >\n") # move on to next stage
    BALROG() # minigame with the BALROG
    
def BALROG(): 
    """
    DOCSTRING:
    --------------------------------------------------------------
    Calling this function will begin the minigame with the BALROG
    No arguments needed. 
    --------------------------------------------------------------
    """
    global power # global variable 
    print("""As you walk through the dark tunnels of the mines, you come across an ancient chamber.
Inscribed in the wall you read a story of the dwarves discovering of a powerful and ancient demon, the Balrog.

Suddenly, you hear a loud rumbling noise coming from deeper in the mines.

You are so close to the exit... 
You just need to cross a bridge but right beneath you is the Balrog.
He seems to be coming in and out of his sleep.

""") # introduction to the situation
    input("< Press Enter to continue >\n") 

    if power>0: # check the power variable, if it has been used previously or not
        print("""You still have your ring. 
You might want to use it now and avoid the risk of getting caught by the Balrog while crossing the bridge.""")
        decision = input("""Type \"Ring\" to use your ring and continue to the next stage. 
You can also press ENTER and start the minigame with the Balrog.""").lower()
        if decision == "ring": # player uses the ring
            power-=1 # reduce the global variable 
            print("""Wise choice! You put on the ring and quickly cross the bridge while invisible.
You made it out of the mines and continue your journey to Mordor""")
            input("< Press Enter to continue to the next stage >\n")
            STAGE3()
    else: # player decided not to use the ring or had an invalid input
        next

# introduce the minigame with the BALROG
    print("""--------------------------------------------------------------------------------------------
In order to cross the bridge you will have to play a minigame with a series of rounds!

Each round the Balrog will either:              Your response should be:

1) Fall asleep                              |   Press "1" -- Move forward    
2) Open his eyes to look for you            |   Press "2" -- Hide behind a Rock                                      
3) Wake up and listen for steps             |   Press "3" -- Stay still

If you mess up once, the Balrog will kill you!
In order to move to the next stage you need to "Move forward" 3 times.

Remember:
Only when he is asleep you can make progress and cross the bridge.
""")
    input("\n< Press Enter to start the minigame >\n") # input before the game starts to give the player time to read
    balrog_actions = ["wakes up and listens for steps", "opens his eyes to look for you", "falls asleep"] # list of actions

    # progression meter for bilbo 
    progression = 0
    while progression<3: # loop until progression has been made 3 times
        # randomly choose one action from balrog_actions list 
        action = random.choice(balrog_actions)
        print(f"\nThe Balrog {action}.\n") # print the random choice

        if action == "falls asleep": 
            # ask user to input 1/2/3 for "walk", "hide", or "stay still" 
            try: # try and except to control the VALUE ERROR that can happen with inputs that require integers
                user_input = int(input("Do you want to (1) Walk, (2) Hide or (3) Stay Still?\n"))

                if user_input == 1: # walk 
                    print("You cautiously take a few steps forward.")  
                    # increase progression meter by 1  
                    progression += 1  

                    if progression == 3: # Bilbo moves on after 3 x success  
                        print("You have successfully crossed the bridge to safety! You move on in your journey to Mordor")  
                        input("\n< Press Enter to continue to the next stage >\n")
                        STAGE3() # move on to the next stage
                        break  

                elif user_input == 2: # hide 
                    print("You quickly hide behind a nearby rock. You did't make any progess when you could have...")  

                elif user_input == 3: # stay still 
                    print("You remain as still as possible. You did't make any progess when you could have...")     
            
            except ValueError : # controlling input 
                print("Invalid input! Please enter either 1, 2 or 3")
        
        elif action == "opens his eyes to look for you": 
            # ask user to input 1/2/3 for "walk", "hide", or "stay still" 
            try: # try and except to control the VALUE ERROR that can happen with inputs that require integers
                user_input = int(input("Do you want to (1) Walk, (2) Hide or (3) Stay Still?\n"))

                if user_input == 1: # walk 
                    print("You cautiously take a few steps forward. The Balrog roars loudly! He saw you move! ")  
                    FAIL(2_1)

                elif user_input == 2: # hide 
                    print("You quickly hide behind a nearby rock. You made the right choice! The Balrog didn't notice you!")  

                elif user_input == 3: # stay still 
                    print("You remain as still as possible. The Balrog roars loudly! He spots you in the middle of the bridge!")  
                    FAIL(2_1)

            except ValueError: # controlling input 
                print("Invalid input! Please enter either 1, 2 or 3")

        elif action == "wakes up and listens for steps": 
            # ask user to input 1/2/3 for "walk", "hide", or "stay still" 
            try: # try and except to control the VALUE ERROR that can happen with inputs that require integers
                user_input = int(input("Do you want to (1) Walk, (2) Hide or (3) Stay Still?\n"))

                if user_input == 1: # walk 
                    print("You cautiously take a few steps forward. The Balrog roars loudly! He heard you move! ")  
                    FAIL(2_1)

                elif user_input == 2: # hide 
                    print("You quickly hide behind a nearby rock. The Balrog roars loudly! He heard you hiding behind the rock! ")  
                    FAIL(2_1)

                elif user_input == 3: # stay still 
                    print("You remain as still as possible. You made the right choice! The Balrog didn't notice you!")  

            except ValueError: # controlling input 
                print("Invalid input! Please enter either 1, 2 or 3")  

def STAGE3(): 
    """
    DOCSTRING:
    --------------------------------------------------------------
    Calling this function will begin Stage 3: "Mordor"
    No arguments needed. 
    --------------------------------------------------------------
    """
# introduction to the next stage
    print("""
-----------------------------------------------------------------------------------------------
Welcome to Mordor you are almost done, so close!
You have to pass through hostile territory to reach Mount Doom, where you can destroy the Ring.
Be careful, the enemy is everywhere...
""")
# explanation of the situation
    print("""You find yourself in front of a huge gate, guarded by the enemy.
Get to the other side to make your way to Mount Doom
""")
    global power 
    if power>0: # if the ring has not been used in a previous stage
        print("""You still have your ring. 
You might want to use it now and avoid the risk of getting caught!""")
        decision = input("""Type \"Ring\" to use your ring and continue to Mount Doom. 
You can also press ENTER to skip \n""").lower()
        if decision == "ring": # player uses the ring, skip the next decision
            power-=1
            print("""Wise choice! You put on the ring and sneak through the gates without beeing seen.
You made it to Mount Doom, the only thing left to do is destroy the ring and save the world.""")
            input("< Press Enter to continue to the final decision >\n")
            FINAL()
    else: # player decides to keep the ring power and move on to the decision
        next
# explain the option
    print("""You are faced with two options:
1) Take a different route into Mount doom 
2) Sneak past the gates 
""")

    choice=0
    while choice not in [1,2]:
        try: # try and except to control the value error with inputs that require integer values 
            choice = int(input("Enter your choice [1 or 2]: "))
            if choice not in [1,2]:
                print("Invalid input. Please try again!")
        except ValueError: 
            print("Invalid input. Please try again!")
            
    if choice == 1: # correct choice
        print("""You chose to take another route around the gates. 
    Trusting your instincts prooves to be a great choice!
    You make it to Mount doom safely.""")
        FINAL() # move on to the final decision
    elif choice == 2: # you get spotted
        FAIL(3)            

def FINAL():
    """
    DOCSTRING:
    --------------------------------------------------------------
    Calling this function will start the final decision of the game
    No arguments needed. 
    --------------------------------------------------------------
    """
    ring_choice = 0
    while ring_choice not in [1,2]:
        try: # try and except to control user input
            ring_choice=  int(input("""You have made it to Mount Doom! Now you must decide to:
1) destroy the ring in the fires of Mount Doom 
2) Keep the ring for yourself
[Choose "1" or "2"] """)) #final decision of the game 
            if ring_choice not in [1,2]:
                print("Invalid input. Please try again!")
        except ValueError: # VALUE ERROR is controlled 
            print("Invalid input. Please try again!")
    if ring_choice == 1:  # player destroys the ring and wins the game
        print("""You have destroyed the ring in the fires of Mount Doom!
Congratulations! You have completed the game! """) 
        AGAIN() # player can choose to play again or not 
    elif ring_choice == 2: # player keeps the ring and looses the game
        FAIL(3_1) # go to fail number 3 


STAGE0()# Start the game

```
